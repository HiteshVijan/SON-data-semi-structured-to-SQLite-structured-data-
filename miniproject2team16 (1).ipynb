Queries:

1. Which company has the highest number of Data Analyst job postings? List the company name, location

```python
%%sql
SELECT company_name, location, COUNT(*) AS count_of_data_analyst
FROM Job_Posting
WHERE title LIKE '%Data Analyst%'
GROUP BY company_name, location
ORDER BY count_of_data_analyst DESC
LIMIT 1
```

2. Find the salary-related attributes for the company with the maximum number of data analyst postings. Do not output missing or 'none' values. Sort the output to have the max salary_avg at the top.

```python
%%sql
SELECT salary_pay, company_name, salary_rate, salary_avg, salary_min, salary_max, salary_hourly, salary_yearly
FROM Job_Posting
WHERE company_name = (
    SELECT company_name
    FROM Job_Posting
    WHERE title LIKE '%Data Analyst%'
    GROUP BY company_name
    ORDER BY COUNT(*) DESC
    LIMIT 1)
 AND salary_avg IS NOT NULL AND salary_avg != 'none'
ORDER BY salary_avg DESC
limit 10;
```

3. Find the two most common platforms for companies to list job postings, and display the count of postings on the platforms

```python
%%sql
SELECT via as platform, COUNT(*) AS posting_count
FROM Job_Posting
GROUP BY platform
ORDER BY posting_count DESC
LIMIT 2;
```

4. How many companies allow Work From Home?

```python
%%sql
SELECT COUNT(DISTINCT company_name) AS num_companies
FROM Job_Posting
WHERE work_from_home = '1';
```

5. How many job postings do the Big 4 firms have, across platforms? Display the count of postings, company name, and the posting platform.


```python
%%sql
SELECT company_name, via as platform, COUNT(*) AS posting_count
FROM Job_Posting
WHERE company_name IN ('KPMG', 'Deloitte', 'PwC', 'EY')
GROUP BY company_name, platform;
```
